```python
# ðŸ“Š Socioeconomic Indicators Visualization from Uploaded ZIP File

import zipfile
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Upload ZIP
from google.colab import files
uploaded = files.upload()  # Upload your 'archive .zip.zip'

# Get ZIP filename automatically
zip_filename = list(uploaded.keys())[0]

# Extract
with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall(".")

# Auto-detect CSV
csv_file = None
for file in os.listdir():
    if file.endswith(".csv"):
        csv_file = file
        print(f"âœ… Found CSV file: {csv_file}")
        break

# Load Data
df = pd.read_csv(csv_file)
print("âœ… Data loaded successfully!")

# Preview
display(df.head())
print(df.info())

# === ðŸ“ˆ VISUALIZATIONS === #

sns.set(style="whitegrid")

# 1. Unemployment Rate Over Years
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='Unemployment Rate (%)', marker='o')
plt.title('Unemployment Rate Over the Years')
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Year')
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Government Social Spending Over the Years
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Year', y='Government Social Spending (% of GDP)', color='skyblue')
plt.title('Government Social Spending Over the Years')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Climate Disasters Per Year
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='Climate Disasters (count)', marker='o', color='red')
plt.title('Climate Disasters Per Year')
plt.ylabel('Disaster Count')
plt.grid(True)
plt.tight_layout()
plt.show()

# 4. Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="YlGnBu")
plt.title("Correlation Matrix of Socioeconomic Indicators")
plt.tight_layout()
plt.show()
```
